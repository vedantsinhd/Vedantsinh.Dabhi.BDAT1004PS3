```python

import pandas as pd

users = pd.read_table(r'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep='|', index_col='user_id')

u_occupation = users.set_index(["occupation"])
      
# step 4 mean age per occupation
mean_age = u_occupation["age"].groupby('occupation').mean()
mean_age
```

```python
# step 5 the Male ratio per occupation and sort it from the most to the least

gender_occupation = users.groupby('gender')['occupation'].value_counts()
sex_ratio = gender_occupation['M']/gender_occupation['F']
print(sex_ratio.sort_values(ascending=False))
```

```python
# step 6 max and min age by occupation
maximum_age =  u_occupation["age"].groupby('occupation').max()
minimum_age =  u_occupation["age"].groupby('occupation').min()
print(maximum_age, minimum_age)
```

```python
# step 7 For each combination of occupation and sex, calculate the mean age
users.groupby(['occupation','gender'])['age'].mean()
```

```python
# step 8 For each occupation present the percentage of women and men

gender_occupation = users.groupby('gender')['occupation'].value_counts()
male_sex_ratio = gender_occupation['M']/(gender_occupation['M'] + gender_occupation['F'])*100
female_sex_ratio = gender_occupation['F']/(gender_occupation['M'] + gender_occupation['F'])*100
print(male_sex_ratio, female_sex_ratio)
```

```python
import pandas as pd
import datetime

url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data"
data = pd.read_csv(url, sep="\s+", parse_dates=[[0,1,2]])

data
```

```python
def date_time(d):
    year = d.year - 100 if d.year > 1989 else d.year
    return datetime.date(year, d.month, d.day)
    
data['Yr_Mo_Dy']  = data['Yr_Mo_Dy'].apply(date_time)

data
```

```python
data["Yr_Mo_Dy"] = pd.to_datetime(data["Yr_Mo_Dy"])
data = data.set_index('Yr_Mo_Dy')
data
```

```python
# Number of non-missing values for each location
data.isnull().sum()
```

```python
data.notnull().sum()
```

```python
# Calculate the mean windspeeds of the windspeeds over all the locations and all the times
data.sum().sum() / data.notna().sum().sum()
```

```python
data.describe(percentiles=[])
```

```python
# create the dataframe
day_st = pd.DataFrame()


day_st['min'] = data.min(axis = 1) 
day_st['max'] = data.max(axis = 1)  
day_st['mean'] = data.mean(axis = 1)
day_st['std'] = data.std(axis = 1) 

day_st.head()
```

```python
# Find the average windspeed in January for each location.
data.loc[data.index.month == 1].mean()
```

```python
# Downsample the record to a yearly frequency for each location
data.groupby(data.index.to_period('A')).mean()
```

```python
# Downsample the record to a monthly frequency for each location.
data.groupby(data.index.to_period('M')).mean()
```

```python
# Downsample the record to a weekly frequency for each location.
data.groupby(data.index.to_period('W')).mean()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
USMD = pd.read_csv('us-marriages-divorces-1867-2014.csv')
USMD.info()
```

```python
USMD.head()
```

```python
years = USMD['Year']
marriages = USMD['Marriages_per_1000']
divorces = USMD['Divorces_per_1000']
USMD = plt.figure(figsize=(16,8))
USMD = plt.plot(years, marriages, label='Marriages')
USMD = plt.plot(years, divorces, label='Divorces')
USMD = plt.title("Number of marriages and divorces per capita in the U.S. between 1867 and 2014")
USMD = plt.xlabel("Years",fontsize=14)
USMD = plt.legend(fontsize = 12, loc = "upper left")
USMD = plt.ylabel("Marriages",fontsize=14)
USMD = plt.grid(True)
USMD
```

```python
data=pd.read_csv('us-marriages-divorces-1867-2014.csv')

year1=data[data['Year']==1900]
year2=data[data['Year']==1950]
year3=data[data['Year']==2000]
all_years=[year1,year2,year3]

new_data=pd.concat(all_years)
new_data.plot(x="Year",y=['Marriages_per_1000','Divorces_per_1000'],kind="bar")
plt.xlabel('Year')
plt.ylabel('Number')
plt.title('Number of marriages and divorces in US per capita between 1900, 1950 and 2000')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

romanEmperors = pd.read_csv('roman-emperor-reigns.csv')
assassinatedEmperors = romanEmperors[romanEmperors['Cause_of_Death'].apply(lambda x: 'assassinated' in x.lower())]

number_assassinated = len(assassinatedEmperors)
deaths = len(romanEmperors) - number_assassinated

names = ['Assassinated', 'Other deaths']
data = [number_assassinated, deaths]
fig = plt.figure(figsize =(10, 7))
plt.pie(data, labels = names)
plt.show()
```
